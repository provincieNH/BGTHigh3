```python
import sys
from rdflib import Graph,URIRef
from gastrodon import RemoteEndpoint,QName,ttl,URIRef,inline
import pandas as pd
```

```python
prefixes=inline("""
    @prefix : <http://bgt.basisregistraties.overheid.nl/bgt/> .
""").graph
```

```python
endpoint=RemoteEndpoint(
    "https://api.labs.kadaster.nl/datasets/bgt-high3/bgt/services/bgt/sparql"
    ,default_graph="https://data.labs.kadaster.nl/bgt-high3/bgt/graphs/default"
    ,prefixes=prefixes
    ,base_uri="http://bgt.basisregistraties.overheid.nl/bgt/"
)
```

```python
q=endpoint.select("""
    SELECT * { ?s ?p ?o .}
    limit 10
""")
q
```

```python
count=endpoint.select("""
    SELECT (COUNT(*) AS ?count) { ?s ?p ?o .}
""").at[0,"count"]
count
```

```python
predicates=endpoint.select("""
    SELECT ?p (COUNT(*) AS ?count) { ?s ?p ?o .} GROUP BY ?p ORDER BY DESC(?count)
""")
predicates
```
