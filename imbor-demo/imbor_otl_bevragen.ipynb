Als de cellen hierboven goed uitgevoerd zijn heb je een lokaal endpoint op een turtle file om sparql queries mee te doen.

De onderstaande functie doet een query om alle eigenschappen bij een object op te vragen, 
waarbij je de label van het object op jan geven ipv de uri

```python
def get_properties(klasse='Boom',lang='nl-NL'):
    #klasse=QName(klasse)
    df=e.select("""
       SELECT ?object ?p ?label ?range {
            ?cls rdfs:label  ?object .
            ?p rdfs:range ?range .
            ?p rdfs:label ?label .
           ?p rdfs:domain ?cls .
           FILTER ( STR(?object) = STR(?_klasse) )    
       }
    """)
    return df.style.highlight_null(null_color='red')
```
